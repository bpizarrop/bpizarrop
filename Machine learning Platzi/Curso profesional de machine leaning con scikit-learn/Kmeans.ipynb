```python
import pandas as pd
from sklearn.cluster import MiniBatchKMeans #se utiliza para que aplique Kmeans a bloques peque√±os consumiendo menos recursos computacionales

import warnings
warnings.simplefilter("ignore")
```

```python
dataset = pd.read_csv('candy.csv')
dataset.head()
```

```python
X = dataset.drop('competitorname', axis = 1)
```

```python
Kmeans = MiniBatchKMeans(n_clusters = 4, batch_size = 8).fit(X)
print('Total de centroides', len(Kmeans.cluster_centers_))
print('='*70)
 
dataset['Cluster'] = Kmeans.predict(X)
```

```python
dataset.head()
```

```python
import seaborn as sns
sns.pairplot(dataset, hue='Cluster')
```
