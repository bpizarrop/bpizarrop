### Medida de riqueza lexica en un texto

* R1 = (Total de palabras unicas) / (Total de palabras)
* R1 = (longitud del vocabulario) / (Longitud del texto)

```python
Vocabulario = sorted(set(text1))
print(Vocabulario[1000:1050]) #imprimimos solo un segmento para revisar el contenido
```

```python
Riqueza_lexica = len(Vocabulario)/ len(text1)
print(Riqueza_lexica)
```

### Automatización del calculo de riqueza lexica

```python
def Riqueza_lexica(texto):
    vocabulario = sorted(set(texto))
    return len(vocabulario) / len(texto)
```

```python
Riqueza_lexica(text1)
```

```python
Riqueza_lexica(text2)
```

#### Porcentaje de la palabra

```python
def porcentaje_palabra(palabra, texto):
    return 100* texto.count(palabra)/ len(texto) 
```

```python
porcentaje_palabra('monster', text1)
```

```python
text1.count('monster')
```

## Estadistica del lenguaje

Se generará un diccionario con el contero de apareición de cada palabra dentro del texto.

```python
dic = {}
for palabra in set(text1):
    dic[palabra] = text1.count(palabra)
dic
```

* el proceso anterior es muy costoso computacionalmente porque cada palabra lo va a buscar al texto una por una.
* Existen otras formas de que sea escalable a un texto mas extenso y menos costoso computacionalmente.

#### Metodo FreqDist de NLTK

```python
fdist = FreqDist(text1)
```

```python
fdist # lo mismo pero usando menos recursos
```

```python
# veremos las 20 palabras mas comunes
fdist.most_common(20)
```

```python
fdist.plot(20)
```
